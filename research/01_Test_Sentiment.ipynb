```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import torch.nn.functional as F

model_name = "distilbert-base-uncased-finetuned-sst-2-english"

model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

tokens = tokenizer.tokenize("I absolutely love how fast the delivery was!")
print(f"Tokens: {tokens}")

token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(f"Token IDs: {token_ids}")

input_ids = tokenizer("I absolutely love how fast the delivery was!")
print(f"Input IDs: {input_ids}")
```

```python
X = ["I absolutely love how fast the delivery was!", "I is a great day"]

batch = tokenizer(X,
                  padding=True,
                  truncation=True,
                  max_length=512,
                  return_tensors='pt')

with torch.no_grad():
     outputs = model(**batch, labels=torch.tensor([1,0])) # labels argument compute the loss in output
     print(outputs)
     predictions = F.softmax(outputs.logits, dim=1)
     print(predictions)
     labels = torch.argmax(predictions, dim=1)
     print(labels)
     labels = [model.config.id2label[label_id] for label_id in labels.tolist()]
     print(labels)
```
